```python
import xml.etree.ElementTree as ET
import os

xml_path = "/kaggle/input/datasets/biditdas06/iu-x-ray/nlm_cxr_reports/ecgen-radiology"

sample_xml = os.listdir(xml_path)[0]

tree = ET.parse(os.path.join(xml_path, sample_xml))
root = tree.getroot()

for elem in root.iter():
    if elem.tag == "AbstractText":
        print("LABEL:", elem.attrib)
        print("TEXT :", elem.text)
        print("-"*100)
```
